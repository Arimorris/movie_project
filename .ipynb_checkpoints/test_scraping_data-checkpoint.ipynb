```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
html_page=requests.get('http://www.boxofficemojo.com/alltime/world/')
```

```python
soup = BeautifulSoup(html_page.content, 'html.parser')
```

```python
soup
```

```python
soup.b
```

```python
links = soup.findAll('b')
```

```python
links
```

```python
links_list=[]
for link in soup.find_all('a'):
    links_list.append(link.get('href'))
```

```python
links_list
```

```python
top_100_movies_links = []
for link in links_list:
    if '/movies/' in link:
        top_100_movies_links.append(link)
```

```python
top_100_movies_links[1:]
```
